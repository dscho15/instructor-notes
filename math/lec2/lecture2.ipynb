```python
import sympy as sp
import bodas
from sympy.plotting import plot
import matplotlib.pyplot as plt
```

```python
""" [summary]
    A little tutorial in Python, where we use it for symbolic manipulations by utilizing 
    the framework sympy
"""

# Define the symbols that are needed
s, t = sp.symbols('s, t')

F_cos = 3 * s/(s**2+3**2)

# Define a laplace function
F = s/(s**2+3**2)

# Do the inverse laplace transformation
f = sp.inverse_laplace_transform(F * F_cos, s, t)

# Plot it
plt.figure(1)
plot(f)
plt.figure(2)
bodas.plot(F)
```

```python
# Define the symbols that are needed
s, t = sp.symbols('s, t')

# Define a laplace function
F = 5/(s+3)

# Do the inverse laplace transformation
f = sp.inverse_laplace_transform(F*F_cos, s, t)

# Plot it
plt.figure(1)
plot(f, (t, -0.01, 5))
plt.figure(2)
bodas.plot(F)
```
