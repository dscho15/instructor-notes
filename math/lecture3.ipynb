```python
import sympy as sp
import numpy as np
from sympy.plotting import plot3d
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
```

This script is primarily used to get a better understanding of saddle points and local minimas, both are often encountered in practise when doing optimization in all sort of forms, i.e. deep learning, convex optimization etc.

The task is, can you make sense of the three plots? Where are the local minimas and the saddle points? Do they have some sort of features?

```python
# Opgave 
# Define the symbols that are needed
x,y = sp.symbols('x y')

f = sp.lambdify((x, y), x*y - x + y, 'numpy')

a = np.linspace(-2, 2, 1000)
b = np.linspace(-2, 2, 1000)
x, y = np.meshgrid(a, b)

fig = plt.figure()
plt.contour(x, y, f(x,y), levels=np.linspace(-5, 5, 100).reshape(-1))
plt.show()
```
